```python
df = pd.read_csv('student_feedback.csv')
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
rating_cols = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]
```

```python
df['overall_rating'] = df[rating_cols].mean(axis=1)
df.head()
```

```python
df.describe()
```

Histograms,

```python
plt.figure(figsize=(8,5))
sns.histplot(df['overall_rating'], bins=10, kde=True)
plt.title('Distribution of Overall Satisfaction Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Count')
plt.show()
```

```python
avg_scores = df[rating_cols].mean().sort_values(ascending=False)
avg_scores
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=avg_scores.values, y=avg_scores.index)
plt.title('Average Rating per Question')
plt.xlabel('Average Score')
plt.ylabel('Feedback Category')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[rating_cols])
plt.xticks(rotation=45)
plt.title('Spread of Ratings for Each Question')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[rating_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Feedback Questions')
plt.show()
```

```python
# Calculate average rating for each category
avg_scores = df[rating_cols].mean().sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=avg_scores.values, y=avg_scores.index)
plt.title('Strengths vs Improvement Areas')
plt.xlabel('Average Rating')
plt.ylabel('Feedback Category')
plt.show()
```

```python
import numpy as np

# Prepare radar chart values
labels = rating_cols
values = df[rating_cols].mean().tolist()
values += values[:1]  # close the loop

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

# Plot
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12)

ax.set_title('Teacher Skill Profile (Radar Chart)', y=1.08)
plt.show()
```
